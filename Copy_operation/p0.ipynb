# üß© Copy Operation in Python---------------

- It is the phenomenon of copying content from one variable into another 
variable. 

- When we assign or copy one variable to another in Python, is known as copy operation

Types of copy operation 

1. General/Normal copy 
2. Shallow copy 
3. Deep copy 

üîπ 1. General / Normal Copy (Assignment Copy)--

- It is the phenomenon of copying content from one variable into 
another variable at the same memory address. 
- When you assign one variable to another using the = operator.
- No new object is created ‚Äî both variables point to the same memory location.
- Any change in one will affect the other. 

Syntax: 

destination_var=source_var


```python
# Example 1

a=[1,2,3]
b=a  # Normal copy
b.append("Happy")
print(a)
print(b)
```

```python
# Example 2

list1 = [1, 2, 3]
list2 = list1       # Normal copy

list2.append(4)

print("list1:", list1)
print("list2:", list2)
```

# üîπ 2. Shallow Copy --

- This type of copy operation allows us to copy the content from one 
variable into another variable by storing it on a different address. 

- A new object is created, but nested (inner) objects are shared between copies.
- Done using copy.copy() from the copy module.

- Changes in inner elements affect both copies.

Syntax: 
- destination_var=source_var.copy() 



```python
# Example 1 

l=[10,23,56] 
l1=l.copy() 
print(l) 
print(l1) 
l.append("hello") 
print(l) 
print(l1)
```

```python
# Example 2 

import copy
list1 = [[10, 20], [30, 40]]
list2 = copy.copy(list1)

list2[0][0] = 99

print("list1:", list1)
print("list2:", list2)
```

 # Drawback: 
‚óè Shallow copy only works for mutable collections.
 
‚óè In presence of a nested collection, we if copy the content 
Present inside any variable into another, and try to modify 
the nested collection then it will be reflected on all of the 
other variables holding the same value copied into them as 
the other one.

# üîπ 3. Deep Copy --

- This type of copy operation was developed to overcome the limitation of 
shallow copy. 
- To perform deep copy we have to follow a syntax 
- Creates a completely independent copy of the object and all its nested objects.
- Done using copy.deepcopy() from the copy module.
- Changes in one object do not affect the other.

Import copy 

- dest_var=copy.deepcopy(source_var)


```python
# Example 1

import copy 
l1=[12,34,["good",'morning']] 
l2=copy.deepcopy(l1) 
l2[2][1]="night" 
print(l1) 
print(l2) 
```

```python
# Example 2 

import copy
list1 = [[10, 20], [30, 40]]
list2 = copy.deepcopy(list1)

list2[0][0] = 99

print("list1:", list1)
print("list2:", list2)
```

| Type of Copy     | Function / Operator | New Object Created | Nested Objects Copied | Changes Affect Original? |
| ---------------- | ------------------- | ------------------ | --------------------- | ------------------------ |
| General / Normal | `=`                 | ‚ùå No               | ‚ùå No                  | ‚úÖ Yes                    |
| Shallow Copy     | `copy.copy()`       | ‚úÖ Yes              | ‚ùå No (shared)         | ‚ö†Ô∏è Yes (for nested)      |
| Deep Copy        | `copy.deepcopy()`   | ‚úÖ Yes              | ‚úÖ Yes                 | ‚ùå No                     |


# ‚öôÔ∏è üî∏ Important Concepts Beyond Normal / Shallow / Deep Copy
üß† 1. id() Function ‚Äî Check Memory Address

To verify whether two variables refer to the same object or not.

a = [1, 2, 3]
b = a
print(id(a), id(b))   # Same address ‚Üí same object


üìò Use this to check if copy is independent or not


üß† 2. Copy Behavior with Immutable vs Mutable Objects

| Type          | Example                | Can be Changed? | Copy Effect                 |
| ------------- | ---------------------- | --------------- | --------------------------- |
| **Immutable** | int, float, str, tuple | ‚ùå No            | Copy behaves like new value |
| **Mutable**   | list, dict, set        | ‚úÖ Yes           | Copy may affect both        |

üß† 3. Copying Collections

Different collections behave differently during copying:

Lists ‚Üí use copy() or slicing [:]

Dictionaries ‚Üí use dict.copy()

Sets ‚Üí use set.copy()

Tuples ‚Üí immutable ‚Üí no need to copy


üß† 4. Copy Module Must Be Imported

import copy is required for copy.copy() and copy.deepcopy().

Without import, Python will show: NameError: name 'copy' is not defined


üß† 5. When to Use Each Copy

| Use Case                                            | Recommended Copy Type    |
| --------------------------------------------------- | ------------------------ |
| Simple list or dict (no nesting)                    | `copy.copy()` or slicing |
| Nested structures (list of lists, dict inside dict) | `copy.deepcopy()`        |
| Temporary reference needed                          | Assignment copy (`=`)    |
