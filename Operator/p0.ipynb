# ⚙️ Operators in Python-

- Operators are the symbols or terms which are used to perform some 
specific operation on operands. 

- Operation: The task that is being done on any data values. 
- Operands:  The data values on which we are performing the task on. 

In python we have a total of 7 types of operators: 

1. Arithmetic operators. 
2. Relational operators 
3. Logical operators 
4. Bitwise operators 
5. Assignment operators 
6. Membership operators 
7. Identity operators

- 1 .Arithmetic operators 
_ These operators are used to perform arithmetic operations on any 
operands. 
- We have 5 types of arithmetic operators 

1. Addition operator(+) 
2. Subtraction operator(-) 
3. Multiplication operator(*) 
4. Division  
1. True division(/) 
2. Floor division(//) 
3. Modulus(%) 
5. Power operator(**) 

| Operator | Description         | Example  |
| -------- | ------------------- | -------- |
| `+`      | Addition            | `a + b`  |
| `-`      | Subtraction         | `a - b`  |
| `*`      | Multiplication      | `a * b`  |
| `/`      | Division            | `a / b`  |
| `//`     | Floor Division      | `a // b` |
| `%`      | Modulus (Remainder) | `a % b`  |
| `**`     | Exponent (Power)    | `a ** b` |



```python
# 1. Addition(+) 
# Ope1 + op2 
# Example:  
a=900 
b=100 
print(a+b) 

# Output: 1000


# 2. Subtraction(-) 
# Ope1 - ope2
# Example:  
a=900
b=100
print(a-b)

# Output: 800

# 3. Multiplication(*) 
# Ope1 * ope2   
# Example:
a=900
b=100
print(a*b)

# Output: 90000  

# 4. Division(/)
# Ope1 / ope2
# Example:
a=900
b=100
print(a/b)

# Output: 9.0

# 5. Modulus(%)
# Ope1 % ope2
# Example:
a=900
b=100
print(a%b)

# Output: 0 

# 6. Exponentiation(**)
# Ope1 ** ope2
# Example:
a=9
b=2
print(a**b)

# Output: 81

# 7. Floor Division(//)
# Ope1 // ope2
# Example:
a=900
b=100
print(a//b)

# Output: 9 


# Note: Ope1 and ope2 are operands.
```
