# Ugly Number (Definition)
An Ugly Number is a positive integer whose prime factors are only 2, 3, or 5.

So, if you factorize the number and find any prime other than 2, 3, or 5,
then it is not an ugly number.


- Example-

| Number | Prime Factors |         Result         |
| :----: | :-----------: | :--------------------: |
|    1   |       —       | ✅ Ugly (by definition) |
|    6   |     2 × 3     |         ✅ Ugly         |
|    8   |   2 × 2 × 2   |         ✅ Ugly         |
|   14   |     2 × 7     |       ❌ Not Ugly       |
|   25   |     5 × 5     |         ✅ Ugly         |



- ⚙️ Algorithm (Steps)

- Input a number n.

- If n <= 0, it’s not ugly.

- Repeatedly divide n by 2, 3, and 5 while it’s divisible.

- If the final value becomes 1 → Ugly Number. Otherwise → Not Ugly Number.


```python
num = int(input("Enter a number: "))
def is_ugly(num):
    if num <= 0:
        return False
    for i in [2, 3, 5]:
        while num % i == 0:

            num //= i
    return num == 1

print(is_ugly(12))
print(is_ugly(14))
print(is_ugly(25))
    
```

```python
def is_ugly(n):
    if n<=0:
        return False
    if n==1:
        return True
    elif n%2==0:
        return is_ugly(n//2)
    elif n%3==0:
        return is_ugly(n//3)
    elif n%5==0:
        return is_ugly(n//5)
    return False

print(is_ugly(12))
print(is_ugly(14))
print(is_ugly(25))
```
