```python
# WAJP to generate numbers from 1 to 10 by using recursion.

def numbers(n):
    if n > 10:   # base condition
        return
    print(n)
    numbers(n + 1)  # recursive call

# function call
numbers(1)


```

```python
# WAJP to generate numbers from 10 to 1 using recursion

def numbers(n):
    if n == 0:   # base condition
        return
    print(n)
    numbers(n - 1)  # recursive call

# function call
numbers(10)
```

```python
# WAJP to generate table of a number using recursion

def print_table(num, i=1):
    if i > 10:   # base condition (table up to 10)
        return
    print(f"{num} x {i} = {num * i}")
    print_table(num, i + 1)   # recursive call

# taking input
n = int(input("Enter a number: "))
print(f"\nTable of {n}:")
print_table(n)

```

```python
# WAJP to generate table of a number using recursion
def mul_tables(n,i=1):
    if i>10:
        return
    print(n,'x',i,'=',n*i)
    mul_tables(n,i+1)
num=int(input("enter the number:"))
mul_tables(num)
```

```python
# wap to print sum of natural numbers

def sum_natural(n):
    if n<=0:
        return 0
    return n+sum_natural(n-1)

print(sum_natural(4))
print(sum_natural(3))   
print(sum_natural(6)) 
```

```python
# sum of square of natural numbers
def cube_sum(n):
    if n<=0:
        return 0
    return n**2 + cube_sum(n-1)
print(cube_sum(3))
print(cube_sum(4))
print(cube_sum(5))
```
