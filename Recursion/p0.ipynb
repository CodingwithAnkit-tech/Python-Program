```python
# WAJP to generate numbers from 1 to 10 by using recursion.

def numbers(n):
    if n > 10:   # base condition
        return
    print(n)
    numbers(n + 1)  # recursive call

# function call
numbers(1)


```

```python
# WAJP to generate numbers from 10 to 1 using recursion

def numbers(n):
    if n == 0:   # base condition
        return
    print(n)
    numbers(n - 1)  # recursive call

# function call
numbers(10)
```

```python
# WAJP to generate table of a number using recursion

def print_table(num, i=1):
    if i > 10:   # base condition (table up to 10)
        return
    print(f"{num} x {i} = {num * i}")
    print_table(num, i + 1)   # recursive call

# taking input
n = int(input("Enter a number: "))
print(f"\nTable of {n}:")
print_table(n)

```

```python
#WAJP to check  a palindrome number by using recursion.

def is_palindrom(n):
    if len(str(n)) <=1:
        return True
    if str(n)[0] != str(n)[-1]:
        return False        
    return is_palindrom(str(n)[1:-1])

print(is_palindrom(121))
print(is_palindrom(22522))
print(is_palindrom(9224299))



```

```python
#WAJP to calculate nth term of Fibonacci series by using recursion.

def fibonacci(n):
    if n==0:
        return 0
    if n==1:

        return 1
    return fibonacci(n-1)+fibonacci(n-2)
# print(fibonacci(0))
# print(fibonacci(1))
# print(fibonacci(5))

n=int(input("enter the number:"))
for i in range(n):
    print(fibonacci(i))
```

```python
# wap to extract only single valued datatype from a tuple taking as input from user

n=eval(input("Enter the tuple:"))
def svdt(n,i=0):
    if i==len(n):
        return[]
    if type(n[i]) in [int,bool,complex]:
        return [n[i]] + svdt(n,i+1)
    return svdt(n,i+1)
print(svdt(n))
```

```python
# wap to check if a number given by user is prime or not.
num = int(input("Enter a number: "))
i = 2
is_prime = True

while i <= num // 2:
    if num % i == 0:
        is_prime = False
        break
    i += 1

if num < 2:
    print("Not Prime")
elif is_prime:
    print("Prime")
else:
    print("Not Prime")
```
