-Packing in Python means putting multiple values into a single variable.
- These values are automatically stored as a tuple.

- There  are two types of packing

1) Tuple Packing / Single packing
2) Dictionary Packing / Double Packing

1) Tuple Packing / Single packing

- It is the phenamena of grouping the individual value in the form of tuple collection to provide security.

2) Dictionary Packing / Double Packing:
- It is a phenamena of grouping the individual key : value pair in the form of dict.
- Dictionary packing usually refers to collecting multiple keyword arguments into a dictionary using **kwargs.
- NOte 
1)  key should follows the rules of identifier
2)  key should be of string datatype only 
3)  Key should be in closed with codes but values must be inclosed  with codes if it is string

- Syntax 

- def fname(**args):
- 
- 
-        SB
-     
- fname (k1=v1,k2=v2,---,kn=vn)

```python
# Example 
def sam(**kw):
    print(kw)
    print(type(kw))
sam(a=1,b=True,c='hii',d =3)
```

```python
# When you use ** in a function parameter, Python packs all keyword arguments into a dictionary.

def student_info(**data):
    print(data)
    print(type(data))

student_info(name="Ankit", age=21, course="B.Tech")
```

```python
#Double asterisk ** is also used for dictionary unpacking.

# Example:
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}

combined = {**dict1, **dict2}
print(combined)
```

# ðŸ”¥ Difference Between *args and **kwargs

| Feature    | `*args`              | `**kwargs`        |
| ---------- | -------------------- | ----------------- |
| Packs into | Tuple                | Dictionary        |
| Accepts    | Positional arguments | Keyword arguments |
| Symbol     | `*`                  | `**`              |
