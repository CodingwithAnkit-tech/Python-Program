# =---------------Arguments-------------------=
- These  are the values pass to the function for the further processing

Typees of Arguments
1) Positional Arguments
2) Default Arguments
3) keyword Arguments
4) Variable length Arguments


1) Positional Arguments
- The arguments which are declarinfg or created in function declaration  are called as position arguments

```python
# Example of Positional Arguments
def sam(a,b):
    print(a)
    print(a)
sam(2,3)
```

```python
# Positional Arguments

# Values are assigned based on position.

def greet(name, age):
    print(name, age)

greet("Ankit", 21)
#Order matters.
```

2) Default Arguments
- they are present in function declaration if the user is passing the the default value then default it will take the value .It will pass the value then defalut value of the variable will be replaced new value


```python
# Example  of  Default Arguments

def sam(a,b=0):
    print(a)
    print(b)
sam(2)
```

```python
# Provide default values in function definition.

def greet(name, age=18):
    print(name, age)

greet("Ankit")       # age = 18 (default)
greet("Rahul", 25)   # age = 25

# âœ” Default arguments must come after non-default parameters.
```

3) keyword Arguments:
- Passing key value pair as the value to the arguments 
- They are present in function call

```python
def sam(**d):
    print(d)
sam(a=3,b=4,c=5)
```

```python
# Arguments passed using parameter names.

def greet(name, age):
    print(name, age)

greet(age=21, name="Ankit")

# âœ” Order does NOT matter.
```

4) Variable length Arguments:
These are the arguments are capable  to taking values from 1 to n numbers of inputs or values .They are present in function declaration.



```python
# Example
def sam(*args,**kwargs):
    print(args)
    print(kwargs)
    print(type(args))
    print(type(kwargs))

sam(2,4,5,6,a=1,b=True)

```
