# Arguments -
- These  are the values pass to the function for the further processing

***  Formal Arguments *** 
- These  are also known as parameter which reciive the value during function call

*** Actual Arguments ***
- These are the actual value pass to the function and present as a function call.

***Types of Arguments***
1) Positional Arguments
2) Keyword Arugments
3) Default Arguments
4) Variable length Arguments


1) Positional Arguments-
- These are present at function declaration and definition
- The length of the form arguments shall be eqal to the Actual Arguments


```python
def main(a,b,c):

    print(a)
    print(b)
    print(c)
main(10,30,40)
```

```python
def person(name,age):
    print("name is:",name)
    print("age is :",age)
person("Aman",22)
```

```python
l = ["hello",90,99+7j]
def unpack(a,b,c):
    print(a,b,c)
unpack(*l)
```

 2) Keyword Arguments-
 

```python
def main(a,b,c):
    print(a)
    print(b)
    print(c)
main(b=20,a=10,c=30)
```

```python
def person(name,age):
    print("name is:",name)
    print("age is :",age)
person(age=10,name="aman")
```

```python
# Default arguments

def main(a=0,b=0,c=0):
    print(a)
    print(b)
    print(c)
main(10,20,30)
```

```python
def person(name,age,email=None):
    print("name is :",name)
    print("age is :",age)
    print("email is :",email)
person("person",16)
```

```python
def person(email,age,name):
    print("name is :",name)
    print("age is :",age)
    print("email is :",email)
person('p@gmail.com',age=12,name='aman')
```

```python

def person(name=None,age=None,email=None):
    print("name is :",name)
    print("age is :",age)
    print("email is :",email)
person(email="py@gmail.com",name="aman",age=89)
```

```python
# Variable length arguments

def main(*args):
    print(args)
main(10,20,'hi','ram')
```

```python
def main(**kwargs):
    print(kwargs)
    
main(a=10,b=20)
```

```python
# # Function to find sum of digits
def sum_of_digits(n):
    return sum(int(d) for d in str(n))

# Main program
print("Prime numbers whose sum of digits is also prime:")
for num in range(2, 200):  # You can change the range as needed
    if is_prime(num):
        digit_sum = sum_of_digits(num)
        if is_prime(digit_sum):
            print(num, end=" ")
```
