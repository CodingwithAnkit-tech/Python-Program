# ğŸ§© Decorators ----

- A decorator in Python is a function that adds extra functionality to another function without changing its original code.
- Decorators are the function which takes another function as input.
- without changing its functionalities.

# syntax

def decorator_name(original_function):
    def wrapper_function():
        # code before
        original_function()
        # code after
    return wrapper_function

@decorator_name
def function_name():
    # main function code

```python
# example of a simple decorator in Python 

def greet_decorator(func):
    def wrapper():
        print("ğŸ˜Š Good Morning!")   # before
        func()
        print("ğŸ‘‹ Have a nice day!")  # after
    return wrapper

@greet_decorator
def say_hello():
    print("Hello, Ankit!")

say_hello()
```

ğŸ’¡ WHY USE DECORATORS?

âœ… To add extra features without changing existing code
âœ… For logging, authentication, timing, etc.
âœ… To make your code cleaner and reusable

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print(f"Execution Time: {end - start:.2f} seconds")
    return wrapper

@timer
def test():
    time.sleep(2)
    print("Task completed!")

test()
```
