```python
# Write a function  to print the length of a list(list is the parametercity)
cities = ["delhi","mumbai","kolkata","chennai", "noida"]
heros = ["ironman","thor","hulk","captain america"]
def print_len(list):
    print(len(list))

print_len(cities)
print_len(heros)

```

```python
#  write  to print the element of a list in a single line
heros = ["ironman","thor","hulk","captain america"]
def print_len(list):
    print(len(list))

def print_list(list):
    for item in list:
        print(item, end=" ")
print_list(heros)
```

```python
n = eval(input("Enter the character :"))
print("string is:",n)
sp_l=[]
for i in n:
    if not i.isalnum():
        print("Special char :",i)
        sp_l.append(i)
print("list of special char :,",sp_l)
```

```python
n = input("Enter the character :")
print("string is:",n)
a=[]
b=[]
c=[]
for i in n:
    if i.isalpha():
        a.append(i)       
    elif i.isdigit():
        b.append(i)
    else:
        c.append(i)
    print("Alphabets :", a)
    print("Digits :", b)
    print("Special :", c)
```
