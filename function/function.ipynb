```python
# Example 3
def add_two_number(a,b,c): # parameters
    result = a+b+c
    print("the sum of two number:",result)
# call function
add_two_number(1,2,4)  # arguments
#-----or-----------
add_two_number(a =1, b=2, c=4)
```

# return statement  

The return statement is used in a functiom to send a result back to the place where the function was called 

When return is executed the function stop running and immediately returns the specified value

```python
# Example 
def add(a,b):
    return a+b
result = add(3,6)
print(result)
```
