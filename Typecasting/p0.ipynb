- üßæ TYPECASTING IN PYTHON---

üß† Definition -

1) Typecasting (Type Conversion) means converting one data type into another.

2) It‚Äôs used when you need to perform operations on incompatible data types.

---------------- OR------------------------

3) It is a phenomenon of converting data from data type to another.

Syntax <required_data_type>(expression)

- Notepoint---

Python supports a wide variety of funtion or method like:int(),float(),str(),ord(),hex(),oct(),typle(),set(),list(),dict(),etc.for the typecasting in python. 



```python
# Example

a = "1"
b = "2"
print(a+b)
```

```python
# using typecasting

a = "1"
b = "2"
print(int(a) + int(b))
```

It can be achived with the help of python built in type conversion function such as int(), float(), hex(), oct(), str(), etc.

```python
# Example--1 

x = "100"
y = int(x)   # convert string to int
z = float(x) # convert string to float

print(y, type(y))
print(z, type(z))
```

```python
#üß† 1Ô∏è‚É£ String ‚Üí Integer / Float

# Convert string containing numbers
x = "100"
y = int(x)         # String to Integer
z = float(x)       # String to Float

print(y, type(y))  # 100 <class 'int'>
print(z, type(z))  # 100.0 <class 'float'>
```

```python
#üß† 2Ô∏è‚É£ Float ‚Üí Integer

a = 45.67
b = int(a)         # Removes decimal part
print(b, type(b))  # 45 <class 'int'>
```

```python
#üß† 3Ô∏è‚É£ Integer / Float ‚Üí String

num = 200
price = 45.9

print(str(num))     # '200'
print(str(price))   # '45.9'
```

```python
name = "Ankit"
age = 22
print(name + " is " + str(age) + " years old.")
```

```python
#4Ô∏è‚É£ String ‚Üí List / Tuple

text = "Python"

print(list(text))   # ['P', 'y', 't', 'h', 'o', 'n']
print(tuple(text))  # ('P', 'y', 't', 'h', 'o', 'n')
```

```python
#5Ô∏è‚É£ List ‚Üí Tuple / Set

nums = [1, 2, 2, 3]

print(tuple(nums))  # (1, 2, 2, 3)
print(set(nums))    # {1, 2, 3}
```

```python
#6Ô∏è‚É£ Tuple ‚Üí List

tp = (10, 20, 30)
lst = list(tp)

print(lst, type(lst))  # [10, 20, 30] <class 'list'>
```

```python
#7Ô∏è‚É£ List of Tuples ‚Üí Dictionary

data = [(1, 'A'), (2, 'B'), (3, 'C')]
dict_data = dict(data)
print(dict_data)        # {1: 'A', 2: 'B', 3: 'C'}
```

```python
#8Ô∏è‚É£ Boolean Conversions

print(int(True))    # 1
print(int(False))   # 0
print(bool(0))      # False
print(bool(25))     # True
```

```python
#9Ô∏è‚É£ Complex Conversion

print(complex(5))        # (5+0j)
print(complex(3, 4))     # (3+4j)
```

# 2Ô∏è‚É£ Implicit Typecasting (Automatic Conversion)

1) According to this one data type is converted into other (automatically), This is called,Implict Typecasting in python. 

2) Python converts small data type to a higher data type to prevent data loss.
 
 --------------------OR--------------------------------------------------------
 
‚û° Done automatically by Python (the interpreter).

‚û° Python automatically converts one data type to another to prevent data loss during operations.

```python
#‚úÖ Example 1:

c = 1.6
d = 8
print(c+d)
```

```python
#‚úÖ Example 2:

a = 10        # Integer
b = 5.5       # Float
c = a + b     # Integer + Float ‚Üí Float
print(c)      # 15.5
print(type(c))# <class 'float'>
```

```python
#‚úÖ Example 3:

x = True
y = 4
z = x + y     # True ‚Üí 1 automatically
print(z)      # 5
print(type(z))# <class 'int'>
```

```python
#Example 1: int + float ‚Üí float

a = 10        # int
b = 2.5       # float
c = a + b
print(c)          # 12.5
print(type(c))    # <class 'float'>
```

```python
#Example 2: bool + int ‚Üí int

x = True      # treated as 1
y = 4         # int
z = x + y
print(z)          # 5
print(type(z))    # <class 'int'>
```

```python
#Example 3: int + complex ‚Üí complex

a = 5         # int
b = 3 + 2j    # complex
c = a + b
print(c)          # (8+2j)
print(type(c))    # <class 'complex'>
```

```python
#Example 4: bool + float ‚Üí float

x = False     # 0
y = 4.8       # float
z = x + y
print(z)          # 4.8
print(type(z))    # <class 'float'>

```

```python
#Example 5: int / int ‚Üí float

a = 7
b = 2
result = a / b
print(result)        # 3.5
print(type(result))  # <class 'float'>
```

```python
#Example 6: Mixing int, float, and complex

a = 5       # int
b = 2.5     # float
c = 3 + 4j  # complex

result = a + b + c
print(result)        # (10.5+4j)
print(type(result))  # <class 'complex'>
```

üîë Type Promotion Order (Low ‚Üí High):


bool ‚Üí int ‚Üí float ‚Üí complex


üß† Key Points:

| Feature        | Explicit Typecasting        | Implicit Typecasting                |
| -------------- | --------------------------- | ----------------------------------- |
| Performed by   | Programmer                  | Python Interpreter                  |
| Control        | Full manual control         | Automatic                           |
| Functions used | int(), float(), str(), etc. | No function ‚Äî happens automatically |
| Example        | `float("5")`                | `5 + 5.5 = 10.5`                    |
