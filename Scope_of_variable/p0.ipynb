# Scope of variable-

- It defines how far we can access and modify the variable.
- Python follows L<E<G<B rules to define scope of variable>
1. L → Local
2. E → Enclosed/Non-local
3. G →Global
4. B →Built in


- Global Scope Variable

-  If we declare any variable in main space is known as global scope of 
variable.
-  We can access global scope variable and main space as well as inside 
the functional space also.
-  We can not modify the global scope variable inside the functional space. , 
in order to do that we have to use global keyword
-  it is a variable which will be declared inside the main space Or
outside the function.
-  They can be accessed both inside and outside of the function.
-  If we want to modify a global variable inside the function, we can
do that but it will not affect the original value.
- So, to modify the original value stored inside the variable we have
to use the keyword “global”.
-  We can access and modify these variables inside the main space
but we can only access them in the method area, but can not
modify them.
-  If we want to modify these variables inside the method area, then
we have to make use of the “global” keyword.
-  “Global” is a keyword used to modify a global variable inside the
method area.



```python
#global
a=10
def main():
    global a # to modify the global variable inside function
    a += 10
    print('inside the functional space:',a)
print('At main space:',a)
main()
```

```python
# global
a=100
b=200
def sum():
 print('inside function')
 print(a,b)
sum()
print("outside func")
print(a,b)
```

- Local Scope Variable-

-  if we declare variable inside a function 
-  we can not access the local scope of variable at main space , it can be access inside the function only
- These are the variables we can access only inside the function and not in mainspace or outside of the function.

```python
# local
a =10
def main():
 b=20
 print('The value of b:',b)
main()
print(b)
```

```python
# local
def sum():
 print('inside function')
 a=340
 b=211
 print(a,b)
sum()
print("outside func")
print(a,b)
```

- Enclosed/non-local Scope Variable --

-  If we declare a variable inside a nested function then it is consider as |
enclosed/non-local scope of variable.
-  We can access this enclosed variable at non-local space and local space.
- • If we try to modify the non-local variable and enclosed variable inside a local  space by default not possible , in order to modify that we have to use non-local  keyword

- Built-in Scope Variable--

-  It is an inbuilt scope which can be access in any module.
-  Ex :- print(),id(),input() etc.
